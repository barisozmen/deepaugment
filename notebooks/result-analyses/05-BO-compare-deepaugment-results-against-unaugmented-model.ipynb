```python
AUG_NOTEBOOK_PATH = "../../reports/experiments/wrn_28_10_2019-02-07_02-04_augmented/wrn_28_10_training_on_cifar10_augmented.csv"
NONAUG_NOTEBOOK_PATH = "../../reports/experiments/wrn_28_10_2019-02-06_07-38_non_augmented/wrn_28_10_training_on_cifar10_non_augmented.csv"
```

```python
import os
import pandas as pd
import numpy as np

%matplotlib inline
import matplotlib.pyplot as plt
import matplotlib.colors as mcolors
```

## Load

```python
aug_df = pd.read_csv(AUG_NOTEBOOK_PATH)
aug_df.head()
```

```python
nonaug_df = pd.read_csv(NONAUG_NOTEBOOK_PATH)
nonaug_df.head()
```

## Plot

```python
plt.figure(figsize=(10,6))

plt.plot(aug_df["epoch"],aug_df["val_acc"], color="salmon")
#plt.plot(aug_df["epoch"],aug_df["acc"], "--",color="salmon")
last_acc = aug_df["val_acc"].values[-1]
plt.text(aug_df.__len__(), last_acc, str(last_acc))

plt.plot(nonaug_df["epoch"],nonaug_df["val_acc"], color="steelblue")
#plt.plot(nonaug_df["epoch"],nonaug_df["acc"], "--",color="steelblue")
last_acc = nonaug_df["val_acc"].values[-1]
plt.text(nonaug_df.__len__(), last_acc, str(last_acc))




plt.xlabel("epoch", fontsize=18)
plt.ylabel("validation accuracy", fontsize=18)
plt.legend(labels=["with aug. data (by deepaugment)","with non-aug. data"], fontsize=14)
```

```python
plt.figure(figsize=(10,6))

plt.plot(aug_df["epoch"],aug_df["val_loss"], color="salmon")
#plt.plot(aug_df["epoch"],aug_df["acc"], "--",color="salmon")
last_loss = aug_df["val_loss"].values[-1]
plt.text(aug_df.__len__(), last_loss, f"{last_loss:.4}")

plt.plot(nonaug_df["epoch"],nonaug_df["val_loss"], color="steelblue")
#plt.plot(nonaug_df["epoch"],nonaug_df["acc"], "--",color="steelblue")
last_loss = nonaug_df["val_loss"].values[-1]
plt.text(aug_df.__len__(), last_loss, f"{last_loss:.4}" )

plt.xlabel("epoch", fontsize=18)
plt.ylabel("validation loss", fontsize=18)
plt.legend(labels=["with aug. data (by deepaugment)","with non-aug. data"], fontsize=14)
```
